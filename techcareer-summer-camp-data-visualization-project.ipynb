```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#  üöÄ 50 Startups Veri Seti 

**50 farklƒ± startup ≈üirketinin** √ße≈üitli harcamalarƒ±nƒ± ve k√¢rlƒ±lƒ±klarƒ±nƒ± i√ßermektedir.  

 **ƒ∞√ßerik:**  

-  **R&D Spend (Ar-Ge Harcamasƒ±)**: ≈ûirketin ara≈ütƒ±rma ve geli≈ütirme (Ar-Ge) i√ßin harcadƒ±ƒüƒ± tutar.  

-  **Administration (Y√∂netim Harcamasƒ±)**: ≈ûirketin y√∂netim giderleri i√ßin harcadƒ±ƒüƒ± tutar.  

-  **Marketing Spend (Pazarlama Harcamasƒ±)**: ≈ûirketin pazarlama ve reklam faaliyetleri i√ßin harcadƒ±ƒüƒ± tutar.  

-  **State (Eyalet)**: ≈ûirketin faaliyet g√∂sterdiƒüi eyalet (√∂rneƒüin, New York, California, Florida).  

-  **Profit (K√¢r)**: ≈ûirketin elde ettiƒüi toplam k√¢r.  

*Bu veri seti, startuplarƒ±n √ße≈üitli harcama kalemleri ile k√¢rlƒ±lƒ±klarƒ± arasƒ±ndaki ili≈ükileri analiz etmek i√ßin kullanƒ±labilir.*  
*√ñrneƒüin, Ar-Ge veya pazarlama harcamalarƒ±nƒ±n k√¢rlƒ±lƒ±k √ºzerindeki etkisini incelemek i√ßin uygun bir veri setidir.*


```python
data = pd.read_csv("/kaggle/input/50-startups/50_Startups.csv")
data.head(10)
```

```python
# Mini EDA
print("üîé Veri Boyutu:", data.shape)
print("\n Veri Tipleri ve Eksik Deƒüerler:")
print(data.info())
print("\n Temel ƒ∞statistikler:")
print(data.describe().T)
```

# üéØ G√∂revler

1. **R&D Harcamasƒ± ve K√¢r Arasƒ±ndaki ƒ∞li≈üki (Scatter Plot): Ar-Ge harcamalarƒ± ile k√¢r** arasƒ±ndaki ili≈ükiyi g√∂steren bir **daƒüƒ±lƒ±m** grafiƒüini √ßiziniz.

2. **Y√∂netim Harcamalarƒ± ve K√¢r Arasƒ±ndaki ƒ∞li≈üki (Scatter Plot): Y√∂netim harcamalarƒ± ile k√¢r** arasƒ±ndaki ili≈ükiyi g√∂steren bir **daƒüƒ±lƒ±m** grafiƒüini √ßiziniz.

3. **Eyaletlere G√∂re Ortalama K√¢r (Bar Chart): Farklƒ± eyaletlerdeki startuplarƒ±n ortalama k√¢rlarƒ±nƒ±** kar≈üƒ±la≈ütƒ±ran bir **√ßubuk** grafik √ßiziniz.

4. **Harcama T√ºrlerinin Kar≈üƒ±la≈ütƒ±rmasƒ± (Boxplot): R&D, y√∂netim ve pazarlama harcamalarƒ±nƒ±n** daƒüƒ±lƒ±mƒ±nƒ± kar≈üƒ±la≈ütƒ±ran bir **kutu** grafiƒüini √ßiziniz.

# 1. G√∂rev: R&D Harcamasƒ± ve K√¢r Arasƒ±ndaki ƒ∞li≈üki (Scatter Plot)

```python
plt.figure(figsize=(12,7))
plt.scatter(data["R&D Spend"], data["Profit"], c=data["R&D Spend"], cmap="viridis", alpha=0.7, edgecolor="k")
plt.colorbar(label="Profit ")
plt.xlabel("R&D Spend")
plt.ylabel("Profit")
plt.title("R&D Harcamasƒ± & K√¢r Arasƒ±ndaki ƒ∞li≈üki")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

# 2. G√∂rev: Y√∂netim Harcamalarƒ± ve K√¢r Arasƒ±ndaki ƒ∞li≈üki (Scatter Plot)

```python
plt.figure(figsize=(12,7))
plt.scatter(data["Administration"], data["Profit"], c=data["Administration"], cmap="plasma", alpha=0.7, edgecolor="k")
plt.colorbar(label="Profit")
plt.xlabel("Administration")
plt.ylabel("Profit")
plt.title("Y√∂netim Harcamasƒ± & K√¢r Arasƒ±ndaki ƒ∞li≈üki")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

# 3. G√∂rev:  Eyaletlere G√∂re Ortalama K√¢r (Bar Chart)

```python
state_profit = data.groupby("State", as_index=False)["Profit"].mean()
colors = sns.color_palette("magma", len(state_profit))
plt.figure(figsize=(7,5))
plt.bar(state_profit["State"], state_profit["Profit"],color=colors, edgecolor="black")
plt.xlabel("Eyalet")
plt.ylabel("Ortalama K√¢r")
plt.title("Eyaletlere G√∂re Ortalama K√¢r")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

# 4. G√∂rev: Harcama T√ºrlerinin Kar≈üƒ±la≈ütƒ±rmasƒ± (Boxplot)

```python
plt.figure(figsize=(12,7))
sns.boxplot(data=data[["R&D Spend", "Administration", "Marketing Spend"]],palette="viridis")
plt.ylabel("Harcama Tutarƒ±")
plt.title("Harcama T√ºrlerinin Kar≈üƒ±la≈ütƒ±rmasƒ± ")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```
