# Kütüphaneler

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

#  🚀 50 Startups Veri Seti 

**50 farklı startup şirketinin** çeşitli harcamalarını ve kârlılıklarını içermektedir.  

 **İçerik:**  

-  **R&D Spend (Ar-Ge Harcaması)**: Şirketin araştırma ve geliştirme (Ar-Ge) için harcadığı tutar.  

-  **Administration (Yönetim Harcaması)**: Şirketin yönetim giderleri için harcadığı tutar.  

-  **Marketing Spend (Pazarlama Harcaması)**: Şirketin pazarlama ve reklam faaliyetleri için harcadığı tutar.  

-  **State (Eyalet)**: Şirketin faaliyet gösterdiği eyalet (örneğin, New York, California, Florida).  

-  **Profit (Kâr)**: Şirketin elde ettiği toplam kâr.  

*Bu veri seti, startupların çeşitli harcama kalemleri ile kârlılıkları arasındaki ilişkileri analiz etmek için kullanılabilir.*  
*Örneğin, Ar-Ge veya pazarlama harcamalarının kârlılık üzerindeki etkisini incelemek için uygun bir veri setidir.*


```python
data = pd.read_csv("/kaggle/input/50-startups/50_Startups.csv")
data.head(10)
```

```python
# Mini EDA
print("🔎 Veri Boyutu:", data.shape)
print("\n Veri Tipleri ve Eksik Değerler:")
print(data.info())
print("\n Temel İstatistikler:")
print(data.describe().T)
```

# 🎯 Görevler

1. **R&D Harcaması ve Kâr Arasındaki İlişki (Scatter Plot): Ar-Ge harcamaları ile kâr** arasındaki ilişkiyi gösteren bir **dağılım** grafiğini çiziniz.

2. **Yönetim Harcamaları ve Kâr Arasındaki İlişki (Scatter Plot): Yönetim harcamaları ile kâr** arasındaki ilişkiyi gösteren bir **dağılım** grafiğini çiziniz.

3. **Eyaletlere Göre Ortalama Kâr (Bar Chart): Farklı eyaletlerdeki startupların ortalama kârlarını** karşılaştıran bir **çubuk** grafik çiziniz.

4. **Harcama Türlerinin Karşılaştırması (Boxplot): R&D, yönetim ve pazarlama harcamalarının** dağılımını karşılaştıran bir **kutu** grafiğini çiziniz.

# 1. Görev: R&D Harcaması ve Kâr Arasındaki İlişki (Scatter Plot)

```python
plt.figure(figsize=(12,7))
plt.scatter(data["R&D Spend"], data["Profit"], c=data["R&D Spend"], cmap="viridis", alpha=0.7, edgecolor="k")
plt.colorbar(label="Profit ")
plt.xlabel("R&D Spend")
plt.ylabel("Profit")
plt.title("R&D Harcaması & Kâr Arasındaki İlişki")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

# 2. Görev: Yönetim Harcamaları ve Kâr Arasındaki İlişki (Scatter Plot)

```python
plt.figure(figsize=(12,7))
plt.scatter(data["Administration"], data["Profit"], c=data["Administration"], cmap="plasma", alpha=0.7, edgecolor="k")
plt.colorbar(label="Profit")
plt.xlabel("Administration")
plt.ylabel("Profit")
plt.title("Yönetim Harcaması & Kâr Arasındaki İlişki")
plt.grid(True, linestyle="--", alpha=0.6)
plt.show()
```

# 3. Görev:  Eyaletlere Göre Ortalama Kâr (Bar Chart)

```python
state_profit = data.groupby("State", as_index=False)["Profit"].mean()
colors = sns.color_palette("magma", len(state_profit))
plt.figure(figsize=(7,5))
plt.bar(state_profit["State"], state_profit["Profit"],color=colors, edgecolor="black")
plt.xlabel("Eyalet")
plt.ylabel("Ortalama Kâr")
plt.title("Eyaletlere Göre Ortalama Kâr")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```

# 4. Görev: Harcama Türlerinin Karşılaştırması (Boxplot)

```python
plt.figure(figsize=(12,7))
sns.boxplot(data=data[["R&D Spend", "Administration", "Marketing Spend"]],palette="viridis")
plt.ylabel("Harcama Tutarı")
plt.title("Harcama Türlerinin Karşılaştırması ")
plt.grid(axis="y", linestyle="--", alpha=0.6)
plt.show()
```
