"""

â€¢ Test seti Ã¼zerinde aynÄ± analizleri tekrarlayÄ±n.

```python
# ðŸ“Œ Gerekli KÃ¼tÃ¼phaneler
import random
import statistics
```

```python
# ðŸ“Œ Veri Seti
kitaplar = [
        {"isim": "Veri Bilimi 101", "yazar": "Ali", "tur": "Bilim", "satis": 1200, "yil": 2021},
        {"isim": "Python ile Yapay Zeka", "yazar": "AyÅŸe", "tur": "Bilim", "satis": 950, "yil":2020},
        {"isim": "Ä°statistik Temelleri", "yazar": "Ali", "tur": "Akademik", "satis": 700, "yil": 2019},
        {"isim": "Makine Ã–ÄŸrenmesi", "yazar": "Can", "tur": "Bilim", "satis": 1800, "yil": 2022},
        {"isim": "Veri GÃ¶rselleÅŸtirme", "yazar": "Deniz", "tur": "Sanat", "satis": 400, "yil": 2018},
        {"isim": "Matematiksel Modelleme", "yazar": "Ali", "tur": "Akademik", "satis": 1500,"yil": 2021},
        {"isim": "Bilgi Toplumu", "yazar": "AyÅŸe", "tur": "Sosyal", "satis": 600, "yil": 2022}
        ]
```

```python
# ðŸ“Œ 1. Fonksiyonlar
def en_cok_satan(kitap_listesi):
    return max(kitap_listesi, key=lambda k: k['satis'])

def yazar_satislari(kitap_listesi):
    satis_dict = {}
    for k in kitap_listesi:
        satis_dict[k['yazar']] = satis_dict.get(k['yazar'], 0) + k['satis']
    return satis_dict

def tur_satislari(kitap_listesi):
    tur_dict = {}
    for k in kitap_listesi:
        tur_dict[k['tur']] = tur_dict.get(k['tur'], 0) + k['satis']
    return tur_dict
```

```python
# ðŸ“Œ 2. Liste ve KÃ¼me Ä°ÅŸlemleri
turler = set(k['tur'] for k in kitaplar)
satis_1000_fazla = [k['isim'] for k in kitaplar if k['satis'] >= 1000]
```

```python
# ðŸ“Œ 3. Lambda / Filter / Map / Sorted
kitap_2020_sonrasi = list(map(lambda k: k['isim'], filter(lambda k: k['yil'] >= 2020, kitaplar)))
satis_10_artis = list(map(lambda k: round(k['satis']*1.10,2), kitaplar))
performans_siralamasi = sorted(kitaplar, key=lambda k: k['satis'], reverse=True)
```

```python
# ðŸ“Œ 4. Genel Ä°statistiksel Analiz
satislar = [k["satis"] for k in kitaplar]
ortalama_satis = statistics.mean(satislar)
std_sapma = statistics.stdev(satislar)
tur_satis = tur_satislari(kitaplar)
en_cok_satan_tur = max(tur_satis, key=tur_satis.get)
```

```python
# ðŸ“Œ 5. Train/Test (%70 train, %30 test)
random.seed(42)  
kitaplar_shuffle = kitaplar.copy()
random.shuffle(kitaplar_shuffle)

train_size = int(0.7 * len(kitaplar_shuffle))
train_set = kitaplar_shuffle[:train_size]
test_set = kitaplar_shuffle[train_size:]

# Train analizleri
train_satislar = [k['satis'] for k in train_set]
train_ortalama = statistics.mean(train_satislar)
train_std = statistics.stdev(train_satislar)
train_yazar = yazar_satislari(train_set)
train_tur = tur_satislari(train_set)

# Test analizleri
test_satislar = [k['satis'] for k in test_set]
test_ortalama = statistics.mean(test_satislar)
test_std = statistics.stdev(test_satislar) if len(test_satislar) > 1 else 0
test_yazar = yazar_satislari(test_set)
test_tur = tur_satislari(test_set)

# KarÅŸÄ±laÅŸtÄ±rma
genel_ortalama = statistics.mean(list(train_yazar.values()))
test_ust_ortalama = [k['isim'] for k in test_set if k['satis'] > genel_ortalama]
```

```python
# ðŸ“Œ Analiz SonuÃ§larÄ±
print("~ ~ ~ ~ ~ KÄ°TAP SATIÅž ANALÄ°Z SÄ°STEMÄ° ~ ~ ~ ~ ~\n")

print("1. TEMEL ANALÄ°Z")
print("----------------------------------------")
print("En Ã§ok satan kitap:", en_cok_satan(kitaplar)['isim'], f"({en_cok_satan(kitaplar)['satis']} adet)")

print("\nYazar satÄ±ÅŸlarÄ±:")
for yazar, satis in yazar_satislari(kitaplar).items():
    print("*", yazar, ":", satis)

print("\nKitap tÃ¼rleri:")
for tur in turler:
    print("*", tur)

print("\n1000+ satÄ±ÅŸ yapan kitaplar:")
for k in satis_1000_fazla:
    print("*", k)

print("\n2. FONKSÄ°YONEL ANALÄ°Z")
print("----------------------------------------")
print("2020 sonrasÄ± kitaplar:")
for k in kitap_2020_sonrasi:
    print("*", k)

print("\n%10 artÄ±ÅŸ:")
for s in satis_10_artis:
    print("*", s)

print("\nPerformans sÄ±ralamasÄ±:")
for idx, k in enumerate(performans_siralamasi, 1):
    print(f"* {idx}. {k['isim']}: {k['satis']} adet")

print("\n3. Ä°STATÄ°STÄ°KSEL ANALÄ°Z")
print("----------------------------------------")
print("Ortalama satÄ±ÅŸ:", round(ortalama_satis,2))
print("Standart sapma:", round(std_sapma,2))
print("En Ã§ok satan tÃ¼r:", en_cok_satan_tur, f"({tur_satis[en_cok_satan_tur]} toplam satÄ±ÅŸ)")

print("\n4. TRAIN/TEST")
print("----------------------------------------")
print("Train set ortalama satÄ±ÅŸ:", round(train_ortalama,2))
print("Train set standart sapma:", round(train_std,2))

print("\nTrain set yazar satÄ±ÅŸlarÄ±:")
for yazar, satis in train_yazar.items():
    print("*", yazar, ":", satis)

print("\nTrain set tÃ¼r daÄŸÄ±lÄ±mÄ±:")
for tur, satis in train_tur.items():
    print("*", tur, ":", satis)

print("\nTest set ortalama satÄ±ÅŸ:", round(test_ortalama,2))
print("Test set standart sapma:", round(test_std,2))

print("\nTest set yazar satÄ±ÅŸlarÄ±:")
for yazar, satis in test_yazar.items():
    print("*", yazar, ":", satis)

print("\nTest set tÃ¼r daÄŸÄ±lÄ±mÄ±:")
for tur, satis in test_tur.items():
    print("*", tur, ":", satis)

print("\nKarÅŸÄ±laÅŸtÄ±rmalÄ± SonuÃ§lar")
print("----------------------------------------")
print("Genel ortalama:", round(genel_ortalama,2))
print("\nTest setinde ortalamanÄ±n Ã¼zerinde satÄ±ÅŸ yapan kitaplar:")
for k in test_ust_ortalama:
    print("*", k)

# KÄ±sa yorum
print("\nYorum")
print("----------------------------------------")
if test_ortalama > train_ortalama:
    print("* Test setinin ortalama satÄ±ÅŸlarÄ±, train setinden daha yÃ¼ksek Ã§Ä±kmÄ±ÅŸtÄ±r. Modelin genellenebilirliÄŸi olumlu gÃ¶rÃ¼nÃ¼yor.")
else:
    print("* Train setinin ortalama satÄ±ÅŸlarÄ± daha yÃ¼ksek. Test setinde performans dÃ¼ÅŸmÃ¼ÅŸ olabilir.")
print("* TÃ¼r ve yazar daÄŸÄ±lÄ±mlarÄ±nda benzerlik/dengesizlik gÃ¶zlenerek daha derin analiz yapÄ±labilir.")
```
