# 🌍 Country Veri Seti

Çeşitli ülkelerle ilgili bir dizi **demografik**, **ekonomik** ve **coğrafi** veriyi içermektedir.  
    
Tabloda:  
- Her bir **satır** → bir ülkeyi temsil eder.  
- Her bir **sütun** → bu ülkelerle ilgili farklı özellikleri gösterir.  

## İçerik

- **Country**: Ülkenin adı  
- **Region**: Ülkenin bulunduğu bölge *(ör. Asya, Doğu Avrupa)*  
- **Population**: Ülkenin toplam nüfusu  
- **Area (sq. mi.)**: Ülkenin yüzölçümü *(mil kare olarak)*  
- **Pop. Density (per sq. mi.)**: Nüfus yoğunluğu *(mil kare başına kişi sayısı)*  
- **Coastline (coast/area ratio)**: Sahil uzunluğunun, ülkenin toplam alanına oranı  
- **Net migration**: Net göç oranı *(giren–çıkan nüfus farkı)*  
- **Infant mortality (per 1000 births)**: Bebek ölüm oranı *(1000 doğum başına)*  
- **GDP ($ per capita)**: Kişi başına düşen Gayri Safi Yurtiçi Hasıla (GSYİH)  
- **Literacy (%)**: Okur-yazarlık oranı  
- **Phones (per 1000)**: Her 1000 kişi başına düşen telefon sayısı  
- **Arable (%)**: Tarıma elverişli arazi yüzdesi  
- **Crops (%)**: Ekilebilir ürünlerin yüzdesi  
- **Other (%)**: Diğer arazi kullanımı yüzdesi  
- **Climate**: Ülkenin iklim kategorisi *(numerik bir değer)*  
- **Birthrate**: Doğum oranı  
- **Deathrate**: Ölüm oranı  
- **Agriculture**: Tarım sektörünün ekonomideki payı  
- **Industry**: Sanayi sektörünün ekonomideki payı  
- **Service**: Hizmet sektörünün ekonomideki payı

## 1. Görev: Nüfusa Göre Azalan Sırada Sıralama

```python
#  En Çok Nüfuslu Ülkeler (İlk 10)
data_sorted_population = data.sort_values(by="Population", ascending=False) 
data_sorted_population.head(10) 
```

```python
# Tüm Ülkeleri Nüfuslarına Göre Azalan Sırada Sıralama
data_sorted_population = data.sort_values(by="Population", ascending=False)
data_sorted_population[["Country", "Population"]].head(227)
```

## 2. Görev: GDP Per Capita Sütununa Göre Artan Sıralama

```python
#  GDP Per Capita Sütununa Göre Artan Sırada İlk 10 Ülke
data_sorted_gdp = data.sort_values(by="GDP ($ per capita)", ascending=True) 
data_sorted_gdp.head(10)
```

```python
# Tüm ülkeler İçin GDP Per Capita Sütununa Göre Artan Sırada Sıralama
data_sorted_gdp = data.sort_values(by="GDP ($ per capita)", ascending=True)
data_sorted_gdp[["Country", "GDP ($ per capita)"]].head(227)
```

## 3. Görev: Population Sütunu 10 Milyonun Üzerinde Olan Ülkeler

```python
# Population Sütunu 10 Milyonun Üzerinde Olan İlk 5 Ülke
data_pop_over_10m = data[data["Population"] > 10_000_000] 
data_pop_over_10m.head()
```

```python
# Population Sütunu 10 Milyonun Üzerinde Olan Ülkeler
data_pop_over_10m = data[data["Population"] > 10_000_000]
data_pop_over_10m[["Country", "Population"]].head(227)
```

## 4. Görev: Literacy (%) Sütununa Göre En Yüksek Okur-Yazarlık Oranına Sahip İlk 5 Ülke

```python
data_sorted_literacy = data.sort_values(by="Literacy (%)", ascending=False)
data_sorted_literacy[["Country", "Literacy (%)"]].head() 
```

## 5. Görev: GDP Per Capita 10.000'in Üzerinde Olan Ülkeler

```python
data_gdp_over_10000 = data[data["GDP ($ per capita)"] > 10_000]
data_gdp_over_10000[["Country", "GDP ($ per capita)"]].head(227)
```
