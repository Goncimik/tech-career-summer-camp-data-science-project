## ƒ∞√ßerik

- **Country**: √úlkenin adƒ±  
- **Region**: √úlkenin bulunduƒüu b√∂lge *(√∂r. Asya, Doƒüu Avrupa)*  
- **Population**: √úlkenin toplam n√ºfusu  
- **Area (sq. mi.)**: √úlkenin y√ºz√∂l√ß√ºm√º *(mil kare olarak)*  
- **Pop. Density (per sq. mi.)**: N√ºfus yoƒüunluƒüu *(mil kare ba≈üƒ±na ki≈üi sayƒ±sƒ±)*  
- **Coastline (coast/area ratio)**: Sahil uzunluƒüunun, √ºlkenin toplam alanƒ±na oranƒ±  
- **Net migration**: Net g√∂√ß oranƒ± *(giren‚Äì√ßƒ±kan n√ºfus farkƒ±)*  
- **Infant mortality (per 1000 births)**: Bebek √∂l√ºm oranƒ± *(1000 doƒüum ba≈üƒ±na)*  
- **GDP ($ per capita)**: Ki≈üi ba≈üƒ±na d√º≈üen Gayri Safi Yurti√ßi Hasƒ±la (GSYƒ∞H)  
- **Literacy (%)**: Okur-yazarlƒ±k oranƒ±  
- **Phones (per 1000)**: Her 1000 ki≈üi ba≈üƒ±na d√º≈üen telefon sayƒ±sƒ±  
- **Arable (%)**: Tarƒ±ma elveri≈üli arazi y√ºzdesi  
- **Crops (%)**: Ekilebilir √ºr√ºnlerin y√ºzdesi  
- **Other (%)**: Diƒüer arazi kullanƒ±mƒ± y√ºzdesi  
- **Climate**: √úlkenin iklim kategorisi *(numerik bir deƒüer)*  
- **Birthrate**: Doƒüum oranƒ±  
- **Deathrate**: √ñl√ºm oranƒ±  
- **Agriculture**: Tarƒ±m sekt√∂r√ºn√ºn ekonomideki payƒ±  
- **Industry**: Sanayi sekt√∂r√ºn√ºn ekonomideki payƒ±  
- **Service**: Hizmet sekt√∂r√ºn√ºn ekonomideki payƒ±

## 1. G√∂rev: N√ºfusa G√∂re Azalan Sƒ±rada Sƒ±ralama

```python
#  En √áok N√ºfuslu √úlkeler (ƒ∞lk 10)
data_sorted_population = data.sort_values(by="Population", ascending=False) 
data_sorted_population.head(10) 
```

```python
# T√ºm √úlkeleri N√ºfuslarƒ±na G√∂re Azalan Sƒ±rada Sƒ±ralama
data_sorted_population = data.sort_values(by="Population", ascending=False)
data_sorted_population[["Country", "Population"]].head(227)
```

## 2. G√∂rev: GDP Per Capita S√ºtununa G√∂re Artan Sƒ±ralama

```python
#  GDP Per Capita S√ºtununa G√∂re Artan Sƒ±rada ƒ∞lk 10 √úlke
data_sorted_gdp = data.sort_values(by="GDP ($ per capita)", ascending=True) 
data_sorted_gdp.head(10)
```

```python
# T√ºm √ºlkeler ƒ∞√ßin GDP Per Capita S√ºtununa G√∂re Artan Sƒ±rada Sƒ±ralama
data_sorted_gdp = data.sort_values(by="GDP ($ per capita)", ascending=True)
data_sorted_gdp[["Country", "GDP ($ per capita)"]].head(227)
```

## 3. G√∂rev: Population S√ºtunu 10 Milyonun √úzerinde Olan √úlkeler

```python
# Population S√ºtunu 10 Milyonun √úzerinde Olan ƒ∞lk 5 √úlke
data_pop_over_10m = data[data["Population"] > 10_000_000] 
data_pop_over_10m.head()
```

```python
# Population S√ºtunu 10 Milyonun √úzerinde Olan √úlkeler
data_pop_over_10m = data[data["Population"] > 10_000_000]
data_pop_over_10m[["Country", "Population"]].head(227)
```
